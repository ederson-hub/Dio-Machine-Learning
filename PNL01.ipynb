```python
from gtts import gTTS

text_to_say = 'How are you doing?'
language = 'en'

tts = gTTS(text=text_to_say, lang=language)
tts.save('output.mp3')
```

```python
from IPython.display import Audio
Audio('output.mp3')
```

```python
french_text = 'Bonjour, comment allez-vous?'
french_tts = gTTS(text=french_text, lang='fr')
french_tts.save('french_output.mp3')
Audio('french_output.mp3')
```
