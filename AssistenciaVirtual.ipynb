```python
# Speech-to-text
import speech_recognition as sr

def speech_to_text():
    """Converts speech from the microphone to text."""
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Adjusting for ambient noise. Please wait...")
        r.adjust_for_ambient_noise(source, duration=5)
        print("Say something!")
        audio = r.listen(source)

    try:
        print("Recognizing...")
        text = r.recognize_google(audio, language='pt-BR')
        print(f"You said: {text}")
        return text
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
        return None
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return None

#speech_to_text()
```

```python
def process_command():
    """Listens for a voice command, converts it to text, and identifies the intention."""
    command_text = speech_to_text()

    if command_text:
        command_text = command_text.lower()

        if "abrir wikipedia" in command_text or "pesquisar na wikipedia" in command_text:
            print("Intenção reconhecida: Abrir Wikipedia")

        elif "abrir youtube" in command_text or "procurar no youtube" in command_text:
            print("Intenção reconhecida: Abrir YouTube")

        elif "encontrar farmácia" in command_text or "farmácia próxima" in command_text:
            print("Intenção reconhecida: Encontrar farmácia próxima")

        else:
            print("Comando não reconhecido. Por favor, tente novamente.")
    else:
        print("Não foi possível reconhecer o comando de voz.")
```

```python
# Implement the action functions
import wikipedia
import pywhatkit

def search_wikipedia(query):
    """Searches Wikipedia for the query and prints a summary."""
    try:
        search_term = query.replace("abrir wikipedia", "").replace("pesquisar na wikipedia", "").strip()
        if search_term:
            wikipedia.set_lang("pt")
            summary = wikipedia.summary(search_term, sentences=2)
            print(f"Aqui está um resumo sobre {search_term}:")
            print(summary)
            text_to_speech(f"Aqui está um resumo sobre {search_term}: {summary}", lang='pt')
        else:
            print("Por favor, especifique o que você gostaria de pesquisar na Wikipedia.")
            text_to_speech("Por favor, especifique o que você gostaria de pesquisar na Wikipedia.", lang='pt')

    except wikipedia.exceptions.DisambiguationError as e:
        print(f"Erro de desambiguação: {e}")
        text_to_speech(f"Erro de desambiguação: {e}", lang='pt')
    except wikipedia.exceptions.PageError as e:
        print(f"Erro de página: {e}")
        text_to_speech(f"Erro de página: {e}", lang='pt')
    except Exception as e:
        print(f"Ocorreu um erro ao pesquisar na Wikipedia: {e}")
        text_to_speech(f"Ocorreu um erro ao pesquisar na Wikipedia: {e}", lang='pt')


def search_youtube(query):
    """Searches and plays a video on YouTube for the query."""
    try:
        search_term = query.replace("abrir youtube", "").replace("procurar no youtube", "").strip()
        if search_term:
            print(f"Abrindo YouTube e procurando por: {search_term}")
            text_to_speech(f"Abrindo YouTube e procurando por {search_term}", lang='pt')
            pywhatkit.playonyt(search_term)
        else:
            print("Por favor, especifique o que você gostaria de procurar no YouTube.")
            text_to_speech("Por favor, especifique o que você gostaria de procurar no YouTube.", lang='pt')
    except Exception as e:
        print(f"Ocorreu um erro ao abrir o YouTube: {e}")
        text_to_speech(f"Ocorreu um erro ao abrir o YouTube: {e}", lang='pt')


def find_nearby_pharmacy(query):
    """Performs a Google search for nearby pharmacies."""
    try:
        print("Procurando por farmácias próximas...")
        text_to_speech("Procurando por farmácias próximas.", lang='pt')
        search_query = "farmácia próxima"
        pywhatkit.search(search_query)
    except Exception as e:
        print(f"Ocorreu um erro ao procurar farmácias: {e}")
        text_to_speech(f"Ocorreu um erro ao procurar farmácias: {e}", lang='pt')


def process_command():
    """Listens for a voice command, converts it to text, and identifies the intention."""
    command_text = speech_to_text()

    if command_text:
        command_text = command_text.lower()

        if "abrir wikipedia" in command_text or "pesquisar na wikipedia" in command_text:
            print("Intenção reconhecida: Abrir Wikipedia")
            search_wikipedia(command_text)
        elif "abrir youtube" in command_text or "procurar no youtube" in command_text:
            print("Intenção reconhecida: Abrir YouTube")
            search_youtube(command_text)
        elif "encontrar farmácia" in command_text or "farmácia próxima" in command_text:
            print("Intenção reconhecida: Encontrar farmácia próxima")
            find_nearby_pharmacy(command_text)
        else:
            print("Comando não reconhecido. Por favor, tente novamente.")
            text_to_speech("Comando não reconhecido. Por favor, tente novamente.", lang='pt')
    else:
        print("Não foi possível reconhecer o comando de voz.")
        text_to_speech("Não foi possível reconhecer o comando de voz.", lang='pt')

# process_command()
```
