```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# read the image
img = cv.imread('lena.jpg')

# prepare a 11x11 averaging filter
kernel = np.ones((11,11), np.float32) / 121

# apply the filter
dst = cv.filter2D(img, -1, kernel)

# change image from RGB space to RGB space
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
dst = cv.cvtColor(dst, cv.COLOR_BGR2RGB)

# display the result
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv.imread('lena.jpg')

# perform laplacian filtering
Laplacian = cv.Laplacian(img,cv.CV_64F)

scobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=5)
scobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)

plt.subplot(2,2,1),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(Laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(scobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(scobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```
