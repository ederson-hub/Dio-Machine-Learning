```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import sys

#if "google.colab" in sys.modules:
#    from google.colab import drive
#    drive.mount('/content/drive')
```

```python
figsize = (10, 10)
```

```python
bgr_im = cv.imread("grama.jpeg")
rgb_im = cv.cvtColor(bgr_im, cv.COLOR_BGR2RGB)
plt.figure(figsize=figsize)
plt.imshow(rgb_im)
plt.title("Grama Original")
plt.show()
```

```python
rgb_green = np.uint8([[[0, 255, 0]]])
hsv_green = cv.cvtColor(rgb_green, cv.COLOR_RGB2HSV)
print(hsv_green)
```

```python
thresholding_schematics = cv.imread("hsv_th_.jpeg")

if thresholding_schematics is not None:
    thresholding_schematics = cv.cvtColor(thresholding_schematics, cv.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    plt.imshow(thresholding_schematics)
    plt.title("Thresholding Schematics")
    plt.show()
else:
    print("Error loading image. Please make sure 'hsv_th.png' is in the correct directory.")
```

```python
hsv_im = cv.cvtColor(rgb_im, cv.COLOR_RGB2HSV)
#Define range of hue and intensity
lower_th = hsv_green - np.array([70, 200, 200])
upper_th = hsv_green + np.array([30, 0, 0])

# Threshold the HSV image
mask = cv.inRange(hsv_im, lower_th, upper_th)

plt.figure(figsize=figsize)
plt.imshow(mask)
plt.title("HSV Image")
plt.show()
```
