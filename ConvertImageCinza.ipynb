```python
from PIL import Image
import numpy as np
from google.colab import files
import io


uploaded = files.upload()

if len(uploaded.keys()) > 0:
    filename = next(iter(uploaded))
    img = Image.open(io.BytesIO(uploaded[filename]))

    # Converter para 255 níveis de cinza
    img_cinza = img.convert('L')
    img_cinza.save('imagem_cinza.jpg')
    print("Imagem em tons de cinza salva como 'imagem_cinza.jpg'")

    # Converter para preto e branco
    threshold = 128
    img_preto_branco = img_cinza.point(lambda p: 255 if p > threshold else 0, '1')
    img_preto_branco.save('imagem_preto_branco.jpg')
    print("Imagem em preto e branco salva como 'imagem_preto_branco.jpg'")

    # Exibir as imagens
    import matplotlib.pyplot as plt

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(img)
    axes[0].set_title('Original Colorida')
    axes[0].axis('off')

    axes[1].imshow(img_cinza, cmap='gray')
    axes[1].set_title('255 Níveis de Cinza')
    axes[1].axis('off')

    axes[2].imshow(img_preto_branco, cmap='gray')
    axes[2].set_title('Preto e Branco')
    axes[2].axis('off')

    plt.show()
else:
    print("Nenhuma imagem foi enviada.")
```
